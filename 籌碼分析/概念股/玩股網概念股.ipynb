```python
import requests
import re
import time
import json
import pickle
```

```python
url = "https://www.wantgoo.com/index/all-industry-of-groups"
headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36 Edg/91.0.864.64"
}
r = requests.get(url, headers= headers)
r
```

```python
industryGroups = json.loads(r.text)
industryGroups
```

```python
# 上市
tseGroups = []
# 上櫃
otcGroups = []
# 興櫃
tpexGroups = []
# 電子股
electronicGroups = []
# 概念股
conceptGroups = []
# 集團股
corporationGroups = []

urlTemplate = "https://www.wantgoo.com/index/%5E{id}/stock-list"
logTemplate = "groupName: {groupName} parse url: {url}"

# stockId stockName marketName marketType eletronics concepts corporations   
stocks = {}

for group in industryGroups:
    if group["groupName"] == "上市":
        tseGroups.append(group)
    elif group["groupName"] == "上櫃":
        otcGroups.append(group)
    elif group["groupName"] == "興櫃":
        tpexGroups.append(group)
    elif group["groupName"] == "電子股":
        electronicGroups.append(group)
    elif group["groupName"] == "概念股":
        conceptGroups.append(group)
    elif group["groupName"] == "集團股":
        corporationGroups.append(group)
        
def parseMarket(groups, groupName):
    for group in groups:
        url = urlTemplate.format(id = group["id"])
        print(logTemplate.format(groupName = group["shortName"], url= url))
        resp = requests.get(url, headers= headers)
        arr = json.loads(resp.text)
        for e in arr:
            idArr = e["id"]
            if idArr not in stocks:
                stocks[idArr] = {
                    "stockName": "",
                    "marketName": "",
                    "marketType": "",
                    "eletronics": [],
                    "concepts": [],
                    "corporations": []
                }
            obj = stocks[idArr]
            if groupName == "電子股":
                obj["eletronics"].append(group["shortName"])
            elif groupName == "概念股":
                obj["concepts"].append(group["shortName"])
            elif groupName == "集團股":
                obj["corporations"].append(group["shortName"])
            else:
                obj["marketName"] = groupName
                obj["marketType"] = group["shortName"]
        time.sleep(1)

            
parseMarket(tseGroups, "上市")
parseMarket(otcGroups, "上櫃")
parseMarket(tpexGroups, "興櫃")
parseMarket(electronicGroups, "電子股")
parseMarket(conceptGroups, "概念股")
parseMarket(corporationGroups, "集團股")

with open('玩股網概念股.pkl', 'wb') as f:
    pickle.dump(stocks, f)
```
